The objective of this practical work was to classify the previously seen Iris data by implementing the K-means and K-mean++ algorithms.


*Is k-means clustering used for supervised or unsupervised classification? Explain your answer.*

K-means clustering is considered an unsupervised classification algorithm, as there is no labeled data given for classification. In order to classify the data set, K-means takes k different random data samples, namely centroids, from the data set, computes the shortest distances from each data point to k centroids. Further, it assigns the data point to the cluster of the centroid to which the distance was the shortest. Next, it adjusts the centroids and the points belonging to the clusters until there are no changes inside of the clusters, or for the given number of iterations.

```python
from kmeans import KMeans
from kmeansplusplus import KMeanPlusPlus
import pandas as pd
```

```python
df = pd.read_csv ('Iris.csv')
df = df.drop(columns = ['Species'])
df
```

```python
feature_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
data = pd.DataFrame(df, columns = feature_columns).to_numpy()
```

*Knowing that we do have a labeled dataset, how many clusters are you going to set in your program?*

As we know from the given data set, that there are three labels (Iris Setosa, Iris Virginica, Iris Versicolor) it makes sense to set the number of cluster to 3, i.e k = 3. 

For the demonstrational purposes we tried the algorithm on k = 3, 4, and 5.


```python
kmplus = KMeanPlusPlus(data, 3)
# kmplus = KMeanPlusPlus(data, 4)
# kmplus = KMeanPlusPlus(data, 3)
kmplus.first_centroid()
kmplus.set_next_centroids()
kmplus.centroids
kmplus.set_clusters()
```

As can be seen from the graphs below, the algorithm does a good job in clustering the irises data set

```python
km3 = KMeans(data, 3)
km3.update()
km3.plot_graphs()
```

Let's visualize the data for k = 4



```python
km4 = KMeans(data, 4)
km4.update()
km4.plot_graphs()
```

Let's visualize the data for k = 5


```python
km5 = KMeans(data, 5)
km5.update()
km5.plot_graphs()
```

Even if there was no prior knowledge of the number of clusters, by visualising the data for different number of k's, it can be observed that k = 3 gives the best partitions.
